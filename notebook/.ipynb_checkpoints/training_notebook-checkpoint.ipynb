```python
def get_data_partitions(data,training_size,val_size,test_size):
    data.shuffle(1000)
    total_data_batch=len(data)
    train_size=round(total_data_batch*training_size)
    remain_size=round(train_size)
    train_ds=data.take(train_size)
    remaining_data=data.skip(train_size)
    val_size=round(remain_size*val_size)
    val_ds=remaining_data.take(val_size)
    test_ds=remaining_data.skip(val_size)
    return train_ds,val_ds,test_ds
```

```python
train_ds,val_ds,test_ds=get_data_partitions(dataset,0.8,0.1,0.1)
```

### chache()
it is used to load the next image until previous image is showing to us during the epochs or looping

### prefetch()
it is used to not the load the image again when the epochs is running or loop is running. so its save time and cpu memory to use again and again

```python
train_ds=train_ds.cache().shuffle(1000).prefetch(buffer_size=tf.data.AUTOTUNE)
test_ds=test_ds.cache().shuffle(1000).prefetch(buffer_size=tf.data.AUTOTUNE)
val_ds=val_ds.cache().shuffle(1000).prefetch(buffer_size=tf.data.AUTOTUNE)
```
