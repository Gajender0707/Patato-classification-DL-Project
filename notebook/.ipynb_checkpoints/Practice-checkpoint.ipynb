# Step 2: Split the data into train,validation and test


```python
TRAIN_SIZE=0.8
TEST_SIZE=0.1
VALIDATION_SIZE=0.1
```

```python
def get_split_data(data,train_size,validation_size,test_size,shuffle=True):
    if shuffle:
        dataset.shuffle(1000)
    total_size=len(data)
    train_size=round(len(data)*train_size)
    train_ds=data.take(train_size)
    val_size=round(len(data)*validation_size)
    remain_data=data.skip(train_size)
    val_ds=remain_data.take(val_size)
    test_ds=remain_data.skip(val_size)
    return train_ds,val_ds,test_ds
    
```

```python
train_ds,val_ds,test_ds=get_split_data(dataset,train_size=TRAIN_SIZE,validation_size=VALIDATION_SIZE,test_size=TEST_SIZE)
```

```python
print(len(train_ds))
print(len(val_ds))
print(len(test_ds))
```

```python
for batch_image,batch_label in val_ds.take(1):
    plt.title(classes[batch_label[0].numpy()])
    plt.imshow(batch_image[0].numpy().astype(int))
```

## step2: cache and prefetch

```python
train_ds=train_ds.cache().shuffle(1000).prefetch(buffer_size=tf.data.AUTOTUNE)
test_ds=test_ds.cache().shuffle(1000).prefetch(buffer_size=tf.data.AUTOTUNE)
val_ds=val_ds.cache().shuffle(1000).prefetch(buffer_size=tf.data.AUTOTUNE)
```
